```python
import pandas as pd
df = pd.read_csv(r'C:\kyc-document-quality\notebooks\outputs\quality_report.csv')

total  = len(df)
passed = (df['overall'] == 'PASS').sum()
failed = (df['overall'] == 'FAIL').sum()

print(f"Total  : {total}")
print(f" PASS : {passed} ({100*passed//total}%)")
print(f" FAIL : {failed} ({100*failed//total}%)")
print()
print("--- Fail reasons ---")
print(f" Blurry      : {df['is_blurry'].sum()}")
print(f" Glare       : {df['has_glare'].sum()}")
print(f" Bad framing : {(~df['good_framing']).sum()}")
```
